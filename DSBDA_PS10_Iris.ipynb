```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Iris.csv", header=None)
df
```

```python
df.columns=['Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'Class']
df
```

```python
df['Class'].unique()
```

```python
setosa_subset = df[df['Class'] == 'Iris-setosa']
versicolor_subset = df[df['Class'] == 'Iris-versicolor']
virginica_subset = df[df['Class'] == 'Iris-virginica']
```

```python
transposed_df = df.transpose()
transposed_df
```

```python
melted_df = pd.melt(df, id_vars=['Class'], var_name='Measurement', value_name='Value')
melted_df
```

```python
wide_df = melted_df.pivot_table(index='Class', columns='Measurement', values='Value')
wide_df
```
