```python
df = pd.read_csv("forestfires.csv")
df
```

```python
bins = [0.0, 1.0, 2.0, 1100.0]
labels = ['Not affected', 'Partially affected', 'Mostly affected']
df['area'] = pd.cut(df['area'], bins=bins, labels=labels, right=False)
```
