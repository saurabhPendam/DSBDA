```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
link='https://www.amazon.in/boAt-Smartwatch-Multiple-Monitoring-Resistance/dp/B096VFND31/ref=sr_1_4?crid=2LO4DWXN1CV83&keywords=smartwatch&qid=1685121447&sprefix=smartwatch%2Caps%2C274&sr=8-4'
```

```python
headers = {
    'User-Agent':'Mozilla/5.0 (Windos NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'
}
page = requests.get(link, headers=headers)
```

```python
page
```

```python
soup = bs(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
names = soup.find_all('span', class_='a-profile-name')
```

```python
names
```

```python
cust_name = []

for i in range(2,len(names)):
  cust_name.append(names[i].get_text())
cust_name
```

```python
len(cust_name)
```

```python
title = soup.find_all('a', class_='review-title-content')
```

```python
title
```

```python
review_title = []
for i in range(0,len(title)):
  review_title.append(title[i].get_text())
review_title
```

```python
review_title[:] = [titles.lstrip('\n') for titles in review_title]
review_title
```

```python
review_title[:] = [titles.rstrip('\n') for titles in review_title]
review_title
```

```python
len(review_title)
```

```python
rating = soup.find_all('i',class_='review-rating')
rating
```

```python
rate = []
for i in range(0,len(rating)):
 rate.append(rating[i].get_text())
rate
```

```python
len(rate)
```

```python
review = soup.find_all("span",{"data-hook":"review-body"})
review
```

```python
review_content = []
for i in range(0,len(review)):
 review_content.append(review[i].get_text())
review_content
```

```python
review_content[:] = [reviews.strip('\n') for reviews in review_content]
review_content
```

```python
len(review_content)
```

```python
df = pd.DataFrame()
df['Customer Name']=cust_name[:8]
df['Review title']=review_title[:8]
df['Ratings']=rate[:8]
df['Reviews']=review_content[:8] 
```

```python
df.head()
```

```python
df.to_csv(r'./SmartWatchReviews.csv',index=True)
```
